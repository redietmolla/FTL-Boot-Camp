```python
from IPython.display import display, Markdown, clear_output
import time

def styled_text(text, color="#000000", size="1em"):
    return f"<span style='color:{color}; font-size:{size}'>{text}</span>"

# Fancy header
display(Markdown("#  BMI Calculator 🏋️"))
display(Markdown("### 🌟 Your Personal Health Assistant 🌟"))
time.sleep(1)
```

```python
try:
    # User input section with animation
    display(Markdown("---"))
    display(Markdown("### 📝 Enter Your Details"))
    weight = float(input("Weight (kg): "))
    height = float(input("Height (m): "))
    
    # Fake calculation animation
    display(Markdown("<br>"))
    for i in range(3):
        display(Markdown(f"🔍 Calculating{'.' * (i+1)}"))
        time.sleep(0.3)
        clear_output(wait=True)
    
    # BMI calculation
    bmi = weight / (height ** 2)
    
    # Result determination
    if bmi < 18.5:
        category = "Underweight 🚩"
        color = "#3498db"
        advice = "📌 Focus on nutrient-rich foods and consult a nutritionist"
    elif 18.5 <= bmi < 25:
        category = "Healthy Weight ✅"
        color = "#2ecc71"
        advice = "🎉 Great job! Maintain your balanced lifestyle!"
    elif 25 <= bmi < 30:
        category = "Overweight ⚠️"
        color = "#f1c40f"
        advice = "🔥 Increase physical activity and monitor diet"
    else:
        category = "Obese ❗"
        color = "#e74c3c"
        advice = "🆘 Consult a healthcare professional immediately"
    
    # Display results
    display(Markdown("---"))
    display(Markdown(f"### 📊 Your Results"))
    display(Markdown(f"**BMI:** {styled_text(f'{bmi:.1f}', color, '2em')}"))
    display(Markdown(f"**Category:** {styled_text(category, color, '1.5em')}"))
    display(Markdown(f"**Advice:** {styled_text(advice, color)}"))
    
   

except ValueError:
    display(Markdown("❌ **Error:** Please enter valid numbers!"))
except ZeroDivisionError:
    display(Markdown("⛔ **Error:** Height cannot be zero!"))

```
